```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data - replace with your actual data
np.random.seed(0)
data1 = np.random.rand(1000).cumsum()
data2 = np.random.rand(1000).cumsum()
data3 = np.random.rand(1000).cumsum()
data4 = np.random.rand(1000).cumsum()
data5 = np.random.rand(1000).cumsum()

# Normalize data to create a CDF
data1 /= data1[-1]
data2 /= data2[-1]
data3 /= data3[-1]
data4 /= data4[-1]
data5 /= data5[-1]

# Sort data
data1.sort()
data2.sort()
data3.sort()
data4.sort()
data5.sort()

# Create a sequence of packet sizes (replace with your actual packet sizes)
packet_sizes = np.linspace(1, 32768, 1000)

# Plot the CDFs
plt.plot(packet_sizes, data1, label='jme 0.86')
plt.plot(packet_sizes, data2, label='lusearch_bug 0.80')
plt.plot(packet_sizes, data3, label='lusearch 0.42')
plt.plot(packet_sizes, data4, label='biojava 0.04')
plt.plot(packet_sizes, data5, label='h2 0.03')

# Customize the plot
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.xlabel('Packet Size')
plt.ylabel('Cumulative density of packets')
plt.title('CDF of Packet Sizes')
plt.legend()
plt.grid(True, which="both", ls="--", linewidth=0.5)

# Show the plot
plt.show()
```
